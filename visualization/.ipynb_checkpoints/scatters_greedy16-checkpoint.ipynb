```python
# Visualizing the 16 chosen features from 68 of >= 2 CIMP+ and 0 CIMP-
# CIMP-H only, combined features
import matplotlib.pyplot as plt
import csv

inpdat = "../featurefilter/68feats.csv"

with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)
datin = open("../greedy/68feats_cover.txt","r")
plots = []
f = []
c = 0
for line in datin:
    X = []
    Y = []
    currmut = (line.split(",")[0]).rstrip()
    ind = listit[0].index(currmut)
    #for each mutation, make a scatterplot
    for w in range(len(listit)):
        if(listit[w][ind] == currmut):
            continue
        X.append(listit[w][0])
        Y.append(listit[w][ind])
        
    plt.figure(figsize=(20,10))
    plt.subplot()
    plt.scatter(X,Y)
    plt.title(currmut)
    plt.xlabel("Samples")
    plt.ylabel("Feature Values")
    plt.xticks(rotation="vertical",fontsize=7)
    c+=1
plt.show()
```
