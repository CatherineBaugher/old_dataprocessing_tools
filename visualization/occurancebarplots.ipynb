```python
import csv

inpdat = "../../3ormoreDATLOW.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

#only look at CIMP-H samps
new = [x for x in listit if str(x[len(listit[0])-1])=="1.0" or x[0] == ""]
listit = new

print("len: ",len(listit))
listit
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

labs = [x[0] for x in listit if x[0] != ""]
y_pos = np.arange(len(labs))

tuple(labs)
```

```python
for y in range(1,len(listit[0])-1): #one plot for each feature
    performance = []
    for x in range(1,len(listit)):
        if(listit[x][0] not in labs):
            continue
        if(float(listit[x][y]) == 1.0):
            performance.append(1)
        else:
            performance.append(0)
    plt.figure(figsize=(20,10))
    plt.subplot()
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, y_pos)
    plt.ylabel('Occurance')
    plt.xlabel('Sample')
    plt.title(listit[0][y])
    
plt.show()
```
