```python
import csv

inpdat = "../../../3ormoreDATLOW.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

#split the two sides by samps with BRAF, and samps without
left = []
right = []
linv = {}
rinv = {}
rtot = {}
ltot = {}
for x in range(1,len(listit)):
    inbraf = False
    if(float(listit[x][len(listit[0])-1]) == 1.0):
        for y in range(1,len(listit[0])-1):
            if(float(listit[x][y]) == 1.0 and listit[0][y].split("_")[0] == "BRAF"):
                inbraf = True
                right.append(listit[x][0])
                if(listit[x][0] not in rinv):
                    rinv[listit[x][0]] = []
    if(not inbraf and float(listit[x][len(listit[0])-1]) == 1.0):
        left.append(listit[x][0])
        if(listit[x][0] not in linv):
            linv[listit[x][0]] = []

print("len of left: ",len(left),"len of right:",len(right))
```

```python
#now inventory all of the mutations that appear in each of these sets
for x in range(1,len(listit)):
    if(listit[x][0] in left or listit[x][0] in right):
        for y in range(1,len(listit[0])-1):
            if(float(listit[x][y]) == 1.0):
                if(listit[x][0] in left):
                    #add mutation to the sample record
                    linv[listit[x][0]].append(listit[0][y])
                    #add to totals of mutations
                    if(listit[0][y] not in ltot):
                        ltot[listit[0][y]] = 0
                    ltot[listit[0][y]] += 1
                if(listit[x][0] in right):
                    rinv[listit[x][0]].append(listit[0][y])
                    if(listit[0][y] not in rtot):
                        rtot[listit[0][y]] = 0
                    rtot[listit[0][y]] += 1

print("RIGHT SIDE MUTATION TOTALS:\n")
ltot = sorted(ltot.items(),key=lambda x: int(x[1]),reverse=True)

ltot
```

```python
print("RIGHT SIDE MUTATION TOTALS:\n")
rtot = sorted(rtot.items(),key=lambda x: int(x[1]),reverse=True)

rtot
```

```python
rinv
```

```python
lhead = []
rhead = []
for mut in ltot:
    lhead.append(mut[0])
for mut in rtot:
    rhead.append(mut[0])
sampsummL = [lhead]
sampsummR = [rhead]

for k in linv:
    newrow = [k]
    for mut in lhead:
        if mut in linv[k]:
            newrow.append("X")
        else:
            newrow.append("")
    sampsummL.append(newrow)
sampsummL

for k in rinv:
    newrow = [k]
    for mut in rhead:
        if mut in rinv[k]:
            newrow.append("X")
        else:
            newrow.append("")
    sampsummR.append(newrow)
sampsummR
```

```python
#export sample summaries
with open("BRAFsampsummaryR.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(sampsummR)
with open("BRAFsampsummaryL.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(sampsummL)
```
