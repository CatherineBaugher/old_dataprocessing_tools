```python
#look at different clusters in our hieratchical diagram
import csv

inpdat = "../../../3ormoreDATLOW.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

split1 = ["TCGA-AA-3516","TCGA-AA-3543","TCGA-AA-A00D","TCGA-A6-2672","TCGA-AA-3664","TCGA-AA-3518","TCGA-AA-3848","TCGA-AA-A03F","TCGA-AA-A020","TCGA-AA-3994","TCGA-AA-3870","TCGA-AA-A00E","TCGA-AG-3575","TCGA-AA-A00A","TCGA-AG-A02X","TCGA-AG-3600","TCGA-AA-3837","TCGA-AA-A029"]
split2 = ["TCGA-AA-3525","TCGA-A6-2676","TCGA-AA-A00J","TCGA-AA-A01P","TCGA-AA-3833","TCGA-AA-3710","TCGA-AA-3672","TCGA-AA-3877","TCGA-AA-3877","TCGA-AA-A022","TCGA-AA-3949","TCGA-AA-3821","TCGA-AA-3947","TCGA-AA-3715"]

print('len of split1: ',len(split1))
print('len of split2: ',len(split2))

counts1 = {}
counts2 = {}
highfeats1 = {}
highfeats2 = {}

dattab1 = ["Mutation","Number Occur in Split","Total TP","Total FP"]
dattab2 = ["Mutation","Number Occur in Split","Total TP","Total FP"]

for x in range(1,len(listit)):
    if(listit[x][0] in split1):
        highfeats1[listit[x][0]] = []
        for y in range(1, len(listit[0])-1):
            if(float(listit[x][y]) == 1.0):
                if(listit[0][y] not in counts1):
                    counts1[listit[0][y]] = 1
                else:
                    counts1[listit[0][y]] += 1
                highfeats1[listit[x][0]].append(listit[0][y])

    if(listit[x][0] in split2):
        highfeats2[listit[x][0]] = []
        for y in range(1, len(listit[0])-1):
            if(float(listit[x][y]) == 1.0):
                if(listit[0][y] not in counts2):
                    counts2[listit[0][y]] = 1
                else:
                    counts2[listit[0][y]] += 1
                highfeats2[listit[x][0]].append(listit[0][y])

    highfeats1 = {k: v for (k,v) in highfeats1.items() if v != []}
    highfeats2 = {k: v for (k,v) in highfeats2.items() if v != []}

def makelist(mydict):
    #build firstrow
    newlist = [[""]]
    for key, value in mydict.items():
        for each in value:
            if each not in newlist[0]:
                newlist[0].append(each)

    for key, value in mydict.items():
        row = [key]
        for y in range(1,len(newlist[0])):
            if(newlist[0][y] in value):
                row.append("X")
            else:
                row.append("")
        newlist.append(row)
    return newlist
            
flist1 = makelist(highfeats1)
flist2 = makelist(highfeats2)

flist2
```

```python
def maketab(mydict):
    dattab = [["Mutation","Number Occur in Split","Total TP","Total FP"]]
    neworder = sorted(mydict.items(), key=lambda x:x[1], reverse=True)
    for item in neworder:
        tp = fp = 0
        nextrow = [item[0], item[1]]
        #find tp and fp
        for x in range(1,len(listit)):
            if(float(listit[x][listit[0].index(item[0])]) == 1.0):
                if(float(listit[x][len(listit[0])-1]) == 1.0):
                    tp += 1
                else:
                    fp += 1
        nextrow.append(tp)
        nextrow.append(fp)
        dattab.append(nextrow)
    return dattab

dattab1 = maketab(counts1)
dattab2 = maketab(counts2)

dattab2
```

```python
with open("split1.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(dattab1)

with open("split2.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(dattab2)
    
with open("splittedfeats1.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(flist1)

with open("splittedfeats2.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(flist2)
```
