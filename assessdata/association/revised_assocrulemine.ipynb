```python
import csv
import itertools

#load the data set
inpdat = "../../../3ormoreDATLOW.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

#filter so dataset is only cimphigh samples
newdata = [listit[0]]
for x in range(1,len(listit)):
    if(float(listit[x][len(listit[0])-1]) == 1.0):
        newdata.append(listit[x])
listit = newdata

len(listit)-1
```

```python
#set frequency threshold
minoccur = 4

#find i-itemsets
numto = 8
#first get k=1 to set up
store = {}
firstset = set()
for y in range(1,len(listit[0])-1):
    count = 0
    thismut = frozenset()
    for x in range(1,len(listit)):
        if(float(listit[x][y]) == 1.0):
            count += 1
    if(count >= minoccur):
        firstset.add(frozenset([listit[0][y]]))
        store[frozenset([listit[0][y]])] = count
print("length of",1,"set:",len(firstset))
sets = [firstset]
#next iterate for each k-itemset
for k in range(1,numto):
    nextset = set()
    for item1 in sets[k-1]:
        for item2 in sets[k-1]:
            if len(item1.union(item2)) == k+1:
                #check if already in
                if(frozenset(item1.union(item2)) in nextset):
                    continue
                #check if union frequent
                count = 0
                for x in range(1,len(listit)):
                    allthere = True
                    for mut in item1.union(item2):
                        if(float(listit[x][listit[0].index(mut)]) != 1.0):
                            allthere = False
                    if(allthere):
                        count += 1
                if(count >= minoccur):
                    nextset.add(frozenset(item1.union(item2)))
                    store[frozenset(item1.union(item2))] = count
    print("length of",k+1,"set:",len(nextset))
    sets.append(nextset)
sets
```

```python
#output results
for k in range(0,len(sets)):
    results = [["mutations in itemset","number of CIMP-High samples with these mutations"]]
    for subs in sets[k]:
        nextrow = []
        collect = []
        for mut in subs:
            collect.append(mut)
        mystr = collect[0]
        for j in range(1,len(collect)):
            mystr += " and "
            mystr += collect[j]
        results.append([mystr,store[subs]])
    #export to csv
    outname = ""
    outname += str(k+1)
    outname += "itemsets.csv"
    with open(outname,"w") as fun:
        writer = csv.writer(fun)
        writer.writerows(results)
        
```
