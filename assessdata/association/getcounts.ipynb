```python
import csv

#load the data
inpdat = "../../../3ormoreDATLOW.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)
inpdat = "./rules.csv"
with open(inpdat,'r') as f:
    it = csv.reader(f)
    rules = list(it)

#make listit only CIMP-High
newdata = [listit[0]]
for x in range(1,len(listit)):
    if(float(listit[x][len(listit[0])-1]) == 1.0):
        newdata.append(listit[x])
listit = newdata

#we want to append a count onto each rule
rules[0].append("count")
for k in range(1,len(rules)):
    muts1 = rules[k][0].split(" and ")
    muts2 = rules[k][1].split(" and ")
    count = 0
    for x in range(1,len(listit)):
        hasall = True
        for y in range(0,len(muts1)):
            if(float(listit[x][listit[0].index(muts1[y])]) != 1.0):
                hasall = False
        for y in range(0,len(muts2)):
            if(float(listit[x][listit[0].index(muts2[y])]) != 1.0):
                hasall = False
        if(hasall):
            count += 1
    rules[k].append(str(count))
rules
```

```python
#output data
with open("./rules.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(rules)
```
