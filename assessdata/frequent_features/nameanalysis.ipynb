```python
import os
import csv
import statistics

inpdat = "../../../coadreadDAT.csv"

with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

storage = {}
for y in range(1,len(listit[0])):
    pieces = listit[0][y].split('_')
    solve = '_'.join(pieces[:-3])
    if solve not in storage:
        storage[solve] = []
    storage[solve].append(listit[0][y])

for item in storage:
    if len(storage[item]) > 1:
        print(storage[item])
```
