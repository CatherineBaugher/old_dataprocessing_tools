```python
from sklearn import manifold, datasets
from statistics import mean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataMatrix = pd.read_csv('../../../3ormoreDATLOW.csv', index_col=0)
dataMatrix.loc[dataMatrix['class'] == 2.0, 'class'] = -1.0

dataMatrix.head()
```

```python

arr = dataMatrix.values
numcols = len(arr[0]) - 1
colheads = list(dataMatrix.index)
collabs = []
for x in range(0, len(arr)):
    if(arr[x][len(arr[0])-1] == 1.0):
#    if(True): #comment out to see only cimp-high
        collabs.append(colheads[x])
print(collabs)
#NOTE!! THIS MAKES IT SO WE ONLY SEE CIMP-HIGH
new = arr[(arr[:,numcols]==1.)]
arr = new
print("len: ",len(arr))


x = arr[:,0:numcols]
y = arr[:,numcols]
for m in range(len(y)):
    if y[m] == -1:
        y[m] = 0
y
```

```python
from sklearn import decomposition
from sklearn import datasets


# Create a regular PCA model 
pca = decomposition.PCA(n_components=2)

# Fit and transform the data to the model
reduced_data_pca = pca.fit_transform(x)

# Inspect the shape
reduced_data_pca.shape

# Print out the data
print(len(reduced_data_pca))
print(reduced_data_pca)
```

```python
def rand_jitter(arr):
    if(len(arr) == 0):
        return arr
    stdev = .01*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev

x = []
y = []
for row in reduced_data_pca:
    x.append(row[0])
    y.append(row[1])
x=np.array(x)
y=np.array(y)
#r squared fns
def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    return m, b
def squared_error(ys_orig,ys_line):
    return sum((ys_line - ys_orig) * (ys_line - ys_orig))
def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, y_mean_line)
    return 1 - (squared_error_regr/squared_error_y_mean)
   
m, b = best_fit_slope_and_intercept(x,y)
regression_line = [(m*i)+b for i in x]

regression_line
```

```python
r_squared = coefficient_of_determination(y,regression_line)
print(r_squared)

plt.txt(0,0,"r")

plt.scatter(x,y,color='#003F72',label='data')
plt.plot(x, regression_line, label='regression line')
plt.legend(loc=4)
plt.show()
```
