```python
from sklearn import manifold, datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches

dataMatrix = pd.read_csv('../../../3ormoreDATLOW.csv', index_col=0)
dataMatrix.loc[dataMatrix['class'] == 2.0, 'class'] = -1.0

dataMatrix.head()
```

```python
# Get colors for braf/kras/both


#note: kras named differently in set of 68 features

#make legend
mypatches = [mpatches.Patch(color='red', label='Has BRAF_GRCh37_7:140453136-140453136 mutation'), mpatches.Patch(color='blue', label='Has KRAS_GRCh37_12:25398284-25398284 mutation'), mpatches.Patch(color='purple', label='Has both mutations'), mpatches.Patch(color='orange', label='Has neither mutations')]

colorsall = []
colorshigh = []
for index, row in dataMatrix.iterrows():
    if(row['BRAF_GRCh37_7:140453136-140453136_Missense-Mutation-SNP-A-A-T_Missense-Mutation-SNP-A-T-T'] == 1 and row['KRAS_GRCh37_12:25398284-25398284_Missense-Mutation-SNP-C-T-T'] == 1):
        if(row['class'] == 1):
            colorshigh.append('purple')
        colorsall.append('purple')
    elif(row['BRAF_GRCh37_7:140453136-140453136_Missense-Mutation-SNP-A-A-T_Missense-Mutation-SNP-A-T-T']):
        if(row['class'] == 1):
            colorshigh.append('red')
        colorsall.append('red')
    elif(row['KRAS_GRCh37_12:25398284-25398284_Missense-Mutation-SNP-C-T-T'] == 1):
        if(row['class'] == 1):
            colorshigh.append('blue')
        colorsall.append('blue')
    else:
        if(row['class'] == 1):
            colorshigh.append('orange')
        colorsall.append('orange')

```

```python

arr = dataMatrix.values
numcols = len(arr[0]) - 1
colheads = list(dataMatrix.index)
collabs = []
for x in range(0, len(arr)):
    if(arr[x][len(arr[0])-1] == 1.0):
#    if(True): #comment to see only cimp-high
        collabs.append(colheads[x])
print(collabs)
#NOTE!! THIS MAKES IT SO WE ONLY SEE CIMP-HIGH
new = arr[(arr[:,numcols]==1.)]
arr = new
print("len: ",len(arr))


x = arr[:,0:numcols]
y = arr[:,numcols]
for m in range(len(y)):
    if y[m] == -1:
        y[m] = 0
y
```

```python
from sklearn import decomposition
from sklearn import datasets


# Create a regular PCA model 
pca = decomposition.PCA(n_components=2)

# Fit and transform the data to the model
reduced_data_pca = pca.fit_transform(x)

# Inspect the shape
reduced_data_pca.shape

# Print out the data
print(len(reduced_data_pca))
print(reduced_data_pca)
```

```python
def rand_jitter(arr):
    if(len(arr) == 0):
        return arr
    stdev = .01*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev
'''colors = ['red', 'blue']
for i in range(len(colors)):
    w = reduced_data_pca[:, 0][y == i] #first principal component
    r = reduced_data_pca[:, 1][y == i] #second principal component
    z = reduced_data_pca[:, 2][y == i] #third principal component
    plt.scatter(rand_jitter(r), rand_jitter(z), c=colors[i])

targ_names = ["Negative", "Positive"]
plt.legend(targ_names, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Second Principal Component')
plt.ylabel('Third Principal Component')
plt.title("PCA Scatter Plot")
#uncomment below to annotate
#for x in range(0,len(reduced_data_pca)):
#    plt.annotate(collabs[x],(reduced_data_pca[x][0],reduced_data_pca[x][1]))
plt.show()'''
```

```python
fig = plt.figure()
ax = Axes3D(fig)

w = []
r = []
z = []

for row in reduced_data_pca:
    w.append(row[0])
    r.append(row[1])
    z.append(row[2])

print(w)
print(rand_jitter(w))    

ax.scatter(rand_jitter(w), rand_jitter(r), rand_jitter(z), c=colorshigh)

ax.legend(handles=mypatches, bbox_to_anchor=(1.05, 1), loc=2)
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
ax.set_title("PCA Scatter Plot - Using 17 Features with 3 or more CIMP-High\nOnly CIMP-High samples are plotted")
```
