```python
#learn about the 2 difficult-to-classify samples
import csv

inpdat = "../../../coadreadDATb4.csv"

with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)

#list our hard-to-find samples
found = {'TCGA-AG-3600': [], 'TCGA-AG-A02X': []}

for x in range(1,len(listit)):
    for y in range(1,len(listit[0])-1):
        if(float(listit[x][y]) == 1.0):
            if(listit[x][0] in found):
                found[listit[x][0]].append(listit[0][y])

found
```
