```python
#Feature selection for implementing Dr. Elnitski's algorithm
import os
import csv
import statistics

inpdat = "../../coadreadDAT.csv"

with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)
    
print("col # of old:",len(listit[0]))
```

```python
endcol = len(listit[0])-1
endrow = len(listit)
ge2 = 0
ge3 = 0
ge4 = 0
ge5 = 0
ge6 = 0
ge7 = 0
ge8 = 0
posscols = [] #only get columns w at least 2 CIMP+ coverage
for y in range(1,endcol):
    count = 0
    neg = 0
    for x in range(1,endrow):
        if(float(listit[x][y]) == 1.0):
            if(float(listit[x][endcol]) == 1.0):
                count += 1
                
    if(count >= 2):
        ge2 += 1
        posscols.append(y)
    if(count >= 3):
        ge3 += 1
    if(count >= 4):
        ge4 += 1
    if(count >= 5):
        ge5 += 1
    if(count >= 6):
        ge6 += 1
    if(count >= 7):
        ge7 += 1
    if(count >= 8):
        ge8 += 1


print(ge2, ge3, ge4, ge5, ge6, ge7, ge8)
```

```python
ranks = open("./rankedacc.txt","w")
scores = {}
ability = {} #use to store FN, FP, TN, TP, accuracy
for y in posscols: #only look at the cols with >= 2 CIMP+ cover
    falseneg = 0
    falsepos = 0
    trueneg = 0
    truepos = 0
    left = []
    right = []
    for x in range(1,endrow): #look at each sample
        if(float(listit[x][y]) == 1.0):
            right.append((listit[x][0],float(listit[x][endcol])))
        elif(float(listit[x][y]) == 0.0):
            left.append((listit[x][0],float(listit[x][endcol])))


    #now that we have a stump, analyze as we would a dead end in dec tree
    #mode rule: if there are equal occurances between -1 and 1, pick 1
    try:
        majl = statistics.mode(x[1] for x in left)
    except:
        majl = 1.0
    try:
        majr = statistics.mode(x[1] for x in right)
    except:
        majr = 1.0


    #now classify all samples under the majority...
    classif = {}
    for x in left:
        classif[x[0]] = majl
    for x in right:
        classif[x[0]] = majr
    #print(classif)
    #now score the results...
    fpsamps = []
    tpsamps = []
    for x in range(1,endrow):
        if(classif[listit[x][0]] == 1.0):
            if(float(listit[x][endcol]) == 1.0):
                truepos += 1
                tpsamps.append(listit[x][0])
            elif(float(listit[x][endcol]) == -1.0):
                falsepos += 1
                tpsamps.append(listit[x][0])
            else:
                print("ERROR1")
        elif(classif[listit[x][0]] == -1.0):
            if(float(listit[x][endcol]) == 1.0):
                falseneg += 1
            elif(float(listit[x][endcol]) == -1.0):
                trueneg += 1
            else:
                print("ERROR2")
        else:
            print("ERROR3")
            print(listit[x][endcol])

    if(falsepos == 0): #only look at these features, no more than 0 FP
        accuracy = 1-((falseneg+falsepos)/(trueneg+falseneg+falsepos+truepos))
        scores[y] = accuracy
        myTup = (truepos,falsepos,trueneg,falseneg)
        ability[y] = myTup
ability
```

```python
keep = [0]

newcsv = [["Feature","True Positives","False Positives","True Negatives","False Negatives"]]
for key,value in sorted(scores.items(), key=lambda v:(-v[1],v[0])):
    if(key in ability.keys()):
        if key != "class":
            keep.append(key)
            ranks.write(listit[0][key] + " : " + str(value) + "\n")
            newcsv.append([listit[0][key],ability[key][0],ability[key][1],ability[key][2],ability[key][3]])

keep.append(len(listit[0])-1)
print("number of new cols: " + str(len(keep)))

newdatset = [[row[c] for c in keep] for row in listit]

with open("filtered_occurrenceandfpALL.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(newdatset)

#output stats of accuracy filter
with open("./statsofaccfilt.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(newcsv)
    
for x in range(0,len(newdatset)):
    print(newdatset[x][len(newdatset[0])-1])
    
ability
```
