```python
#Feature selection for implementing Dr. Elnitski's algorithm
import os
import csv
import statistics

inpdat = "../../coadreadDAT2.csv"

with open(inpdat,'r') as f:
    it = csv.reader(f)
    listit = list(it)
    
print("col # of old:",len(listit[0]))
print("row #:",len(listit))
```

```python
endcol = len(listit[0])-1
endrow = len(listit)

#count feats with 0 cimp-
posscols = [] #only get columns w at least 2 CIMP+ coverage
count = count1 = count2 = 0
for y in range(1,endcol):
    FPcount = 0
    for x in range(1,endrow):
        if(float(listit[x][y]) == 1.0):
            if(float(listit[x][endcol]) == -1.0):
                FPcount += 1
    if(FPcount == 0):
        count += 1
    if(FPcount <= 1):
        count1 += 1
    if(FPcount <= 2):
        count2 += 1
        
print("num feats with 0 CIMP-: " + str(count))
print("num feats with 1 or less CIMP-: " + str(count1))
print("num feats with 2 or less CIMP-: " + str(count2))

posscols = [0] #only get columns w at least 2 CIMP+ coverage
posscols2 = []
posscols3 = []
posscols4 = []
posscols5 = []
posscols6 = []
posscols7 = []
posscols8 = []
posscount2 = posscount3 = posscount4 = posscount5 = posscount6 = posscount7 = posscount8 = 0
for y in range(1,endcol-1):
    TPcount = 0
    FPcount = 0
    for x in range(1,endrow):
        if(float(listit[x][y]) == 1.0):
            if(float(listit[x][endcol]) == 1.0):
                TPcount += 1
            else:
                FPcount += 1
                
    if(TPcount >= 2):
        if(FPcount == 0):
            posscols.append(y)
            posscols2.append(y)
        posscount2 += 1
    if(TPcount >= 3):
        if(FPcount <= 2):
            posscols3.append(y)
        posscount3 += 1
    if(TPcount >= 4):
        if(FPcount <= 2):
            posscols4.append(y)
        posscount4 += 1
    if(TPcount >= 5):
        if(FPcount == 0):
            posscols5.append(y)
        posscount5 += 1
    if(TPcount >= 6):
        if(FPcount == 0):
            posscols6.append(y)
        posscount6 += 1
    if(TPcount >= 7):
        if(FPcount == 0):
            posscols7.append(y)
        posscount7 += 1
    if(TPcount >= 8):
        if(FPcount == 0):
            posscols8.append(y)
        posscount8 += 1
    
print("count 2: " + str(posscount2) + " count 3: " + str(posscount3) + " count 4: " + str(posscount4) + " count 5: " + str(posscount5) + " count 6: " + str(posscount6) + " count 7: " + str(posscount7) + " count 8: " + str(posscount8))
print("inter 2: " + str(len(posscols2)) +  " inter 3: " + str(len(posscols3)) + " inter 4: " + str(len(posscols4)) + " inter 5: " + str(len(posscols5)) + " inter 6: " + str(len(posscols6)) + " inter 7: " + str(len(posscols7)) + " inter 8: " + str(len(posscols8)))
posscols.append(len(listit[0])-1)
posscols
```

```python
newdatset = [[row[c] for c in posscols] for row in listit]
with open("filtered_occurrenceandfpCOUNTS.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(newdatset)
```
