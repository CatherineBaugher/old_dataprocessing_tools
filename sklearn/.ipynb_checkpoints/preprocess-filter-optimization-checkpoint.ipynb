```python
'''
Using scikit-learn library to generate random forest evaluator
w/ custom k-fold (bagged samples) cross-validation
using different constrained info gain filter
'''

#import libraries
import sklearn.ensemble as ensemble
import sklearn.datasets as datset
import pandas as pd
import numpy as np

#load up CSV into dataframe
#we want to do this on many different filtered constraints of info gain
rounds = []
rounds.append(pd.read_csv('../featurefilter/filtered_infogain05.csv', index_col=0))
rounds.append(pd.read_csv('../featurefilter/filtered_infogain025.csv', index_col=0))
rounds.append(pd.read_csv('../featurefilter/filtered_infogain015.csv', index_col=0))
rounds.append(pd.read_csv('../featurefilter/filtered_infogain01.csv', index_col=0))
rounds.append(pd.read_csv('../featurefilter/filtered_infogain005.csv', index_col=0))
rounds.append(pd.read_csv('../featurefilter/filtered_infogain0025.csv', index_col=0))

#iris = datset.load_iris()
#make a pandas dataframe w/ (sample,feature) format
#dataMatrix = pd.DataFrame(iris.data, columns=iris.feature_names)
#view top 5 rows
rounds[5].head()
```

```python
#perform a k-fold split
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_classification
from math import sqrt
import random
#from sklearn.model_selection import cross_val_predict

avgs = []
avgs.append(["",".5%",".25%",".15%",".10%",".05%",".025%"])
#do some number of runs to get better idea of accuracy
for u in range(50):
    newrow = [str(u+1)]
    for dataMatrix in rounds:
        array = dataMatrix.values

        numcols = len(array[0]) - 1

        x = array[:,0:numcols]
        y = array[:,numcols]

        #set params for RF
        #seed = 7
        numtrees = 1000
        numfeats = len(dataMatrix.columns) - 1
        numsamps = len(dataMatrix)
        maxfeats = int(numfeats) #randomly select features of size sqrt(num features)

        kf = KFold(n_splits=10)
        #kf.get_n_splits(x)

        #make our own custom cross validation w 2d matrix
        indices = []
        #get random sampling of training indices...
        for trainIndex, testIndex in kf.split(x):
            newTrainIndex = []
            for j in range(0,len(trainIndex)):
                chosenCol = random.randint(0,len(trainIndex)-1) #pick random trainIndex index
                newTrainIndex.append(trainIndex[chosenCol])
            trainIndex = newTrainIndex
            myTuple = (trainIndex,testIndex)
            indices.append(myTuple)

            #need to fit the model, so describe the classification task first
        J,k = make_classification(
            n_samples = numsamps,
            n_features = numfeats,
            n_informative = 10,
            n_redundant = 0,
            n_repeated = 0,
            n_classes = 2,
            shuffle = False
            )

        #create/train our model
        RFmodel = ensemble.RandomForestClassifier(n_estimators = numtrees, max_features = maxfeats)

        RFmodel.fit(J,k)

        importances = RFmodel.feature_importances_

        results = cross_val_score(RFmodel, x, y, cv = indices)

        for z in range (1, len(results)+1):
            print("Accuracy of fold " + str(z) + ": " + str(results[z-1]))

        print("\nAvg accuracy across folds: " + str(results.mean()))
        newrow.append(results.mean())
        
    avgs.append(newrow)
    


avgs
```

```python
import matplotlib.pyplot as plt
import statistics
#plot the data we just generated
labels = avgs[0][1:]
npavgs = np.array(avgs[1:])
npavgs = npavgs[:, 1:]
plt.figure(figsize=(20,10))
plt.plot(npavgs)
plt.legend(labels)
plt.savefig("filter_threshold_vs_accuracy.png")
plt.show()
plt.close()

contendors = []
for m in range(50):
    peakat = max(npavgs[m])
    for index,v in enumerate(npavgs[m]):
            if v == peakat:
                contendors.append(index)
print("Most peaks: " + labels[statistics.mode(contendors)])
    
```
