```python
#This builds a dataset using new criteria:
#Mutations at the same coordinates are treated as one
#CIMP low is considered CIMP+
import os
import csv

datclinfile = "../../data_clinical.csv"
mutextfile = "../../data_mutations_extended.csv"
with open(datclinfile,'r') as f:
        it = csv.reader(f)
        datclinall = list(it)
with open(mutextfile,'r') as f:
        it = csv.reader(f)
        mutextall = list(it)

presents = {} #map samp to all present mutations
firstrow = [""] #this will list all mutations at the top
#find mutations within various samples, cut off past
for x in range(1,len(mutextall)):
    #only use the "mutation" if it's labelled somatic...
    if(mutextall[x][25] != "Somatic"):
        continue
    pieces = mutextall[x][15].split("-")
    inp = pieces[0] + "-" + pieces[1] + "-" + pieces[2] #sample name
    firstpart = (mutextall[x][0] +"_"+ mutextall[x][3] + "_" +mutextall[x][4] + ":"+ mutextall[x][5] + "-" + mutextall[x][6])
    descriptor = mutextall[x][8] + mutextall[x][9] + "-" + mutextall[x][10] + "-" + mutextall[x][11] + "-" + mutextall[x][12]
    descriptor = descriptor.replace('_','')
    flag = False
    for k in range(0,len(firstrow)):
        if(firstrow[k] == ""):
            continue
        parts = firstrow[k].split("_")
        newparts = parts[0] + '_' + parts[1] + '_' + parts[2]
        if firstpart == newparts:
            #print(firstpart)
            flag = True #feature is already in set
            if(descriptor in parts):
                #this mutation is already in the name
                break
            else:
                #append new mutation type
                firstrow[k] = firstrow[k] + "_" + descriptor
                print(firstrow[k])
                
    if not flag: #we have a new mutation
        firstrow.append(firstpart+"_"+descriptor)

    if inp in presents:
        presents[inp].append(firstpart)
    else:
        presents[inp] = [firstpart]

        
firstrow
```

```python
firstrow.append("class")
numfeats = len(firstrow) - 2
print("num features",str(numfeats))

#find foreground vs. background
newset = [firstrow]
foreback = {}
for x in range(6,len(datclinall)-1):
    if datclinall[x][0] in foreback:
        print("ERROR.") #multiple of same sample
    else:
        if(datclinall[x][7] == "CIMP.H"):
            foreback[datclinall[x][0]] = "1.0"
        else:
            foreback[datclinall[x][0]] = "-1.0"

#attach to new matrix
for key,muts in presents.items():
    nextrow = []
    nextrow.append(key)
    for item in firstrow:
        if(item != "" and item != "class"):
            deconstruct = item.split("_")
            item = (deconstruct[0] + "_" + deconstruct[1] + "_" + deconstruct[2])
            if item in muts:
                nextrow.append("1.0")
            else:
                nextrow.append("0.0")
    nextrow.append(foreback[key])
    newset.append(nextrow)

with open("coadreadDATcomb.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(newset)
```
