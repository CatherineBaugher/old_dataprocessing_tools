```python
#This builds a dataset using new criteria:
#Mutations at the same coordinates are treated as one
#CIMP low is considered CIMP+
import os
import csv

datclinfile = "../../data_clinical.csv"
mutextfile = "../../data_mutations_extended.csv"
with open(datclinfile,'r') as f:
        it = csv.reader(f)
        datclinall = list(it)
with open(mutextfile,'r') as f:
        it = csv.reader(f)
        mutextall = list(it)

storage = {} #store mutations within hugo ids
presents = {} #map samp to all present mutations
firstrow = [""] #this will list all mutations at the top
#find mutations within various samples, cut off past
for x in range(1,len(mutextall)):
    #only use the "mutation" if it's labelled somatic...
    if(mutextall[x][25] != "Somatic"):
        continue
    pieces = mutextall[x][15].split("-")
    inp = pieces[0] + "-" + pieces[1] + "-" + pieces[2] #sample name
    firstpart = mutextall[x][0] #hugo
    descriptor = mutextall[x][3] + "_" +mutextall[x][4] + ":"+ mutextall[x][5] + "-" + mutextall[x][6] +"_"+ mutextall[x][8] + "_" + mutextall[x][9] + "_" + mutextall[x][10] + "_" + mutextall[x][11] + "_" + mutextall[x][12]
    
    if firstpart not in firstrow: #we have a new mutation
        firstrow.append(firstpart)

    #add hugo id to sample
    if inp not in presents:
        presents[inp] = []
    presents[inp].append(firstpart)
    
    #add mutation types to hugo id
    if(firstpart not in storage):
        storage[firstpart] = []
    storage[firstpart].append(firstpart+"_"+descriptor)

        
firstrow
```

```python
firstrow.append("class")
numfeats = len(firstrow) - 2
print("num features",str(numfeats))

#find foreground vs. background
newset = [firstrow]
foreback = {}
for x in range(6,len(datclinall)-1):
    if datclinall[x][0] in foreback:
        print("ERROR.") #multiple of same sample
    else:
        if(datclinall[x][7] == "CIMP.H"):
            foreback[datclinall[x][0]] = "1.0"
        elif(datclinall[x][7] == "CIMP.L"):
            foreback[datclinall[x][0]] = "2.0"
        else:
            foreback[datclinall[x][0]] = "-1.0"

#attach to new matrix
for key,muts in presents.items():
    nextrow = []
    nextrow.append(key)
    for item in firstrow:
        if item == "" or item == "class":
            continue
        if item in muts:
            nextrow.append("1.0")
        else:
            nextrow.append("0.0")
    nextrow.append(foreback[key])
    newset.append(nextrow)

with open("coadreadDAThugocombine.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(newset)
```

```python
#make a new summary of hugo ids and mutations in them
combinationstats = [["HUGO ID","Number of mutations","Mutations within ID","..."]]
for row in sorted(storage, key=lambda k: len(storage[k]), reverse=True):
    newrow = [row,len(storage[row])]
    for mut in storage[row]:
        newrow.append(mut)
    combinationstats.append(newrow)

with open("hugocombinestats.csv","w") as fun:
    writer = csv.writer(fun)
    writer.writerows(combinationstats)

combinationstats
```
