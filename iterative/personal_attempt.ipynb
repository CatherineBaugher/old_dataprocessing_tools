```python
from deap import creator
from deap import base
from deap import tools
from deap import algorithms
import random
import csv
import numpy

#create a fitness fn where we maximize first objective, minimize the second
creator.create("FitnessMulti",base.Fitness,weights=(1.0,-1.0))
#create an individual class which has a fitness attribute of the FitnessMulti we just made
creator.create("Individual",list,fitness=creator.FitnessMulti)

#get dataset
with open("../../selectedfeatures2withcimplow.csv",'r') as f:
    it = csv.reader(f)
    listit = list(it)

#number of individuals
POPNUM = 50
#number of generations
GENNUM = 50    
#number of features
FEATNUM = len(listit[0])-2

listit
```

```python
#the eval fn is just the sum of weights
def evalMax(individual):
    return sum(individual)

#make our toolbox to store fns and arguments
toolbox = base.Toolbox()
toolbox.register("attr_bool",random.randint,0,1)
toolbox.register("individual",tools.initRepeat,creator.Individual,toolbox.attr_bool, n=FEATNUM)
toolbox.register("population",tools.initRepeat,list,toolbox.individual)
#genetic ops
toolbox.register("evaluate",evalMax)
toolbox.register("mate",tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=.05)
toolbox.register("select",tools.selTournament,tournsize=3)

#print(indlist[2])
```

```python
hof = tools.HallOfFame(1)
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", numpy.mean)
stats.register("min", numpy.min)
stats.register("max", numpy.max)
    
pop = toolbox.population(n=POPNUM)
   
pop,logbook = algorithms.eaSimple(pop,toolbox,cxpb=0.5,mutpb=0.2,ngen=GENNUM,stats=stats,halloffame=hof,verbose=True)


print("Best individual is: %s\nwith fitness: %s" % (hof[0],hof[0].fitness))
```
